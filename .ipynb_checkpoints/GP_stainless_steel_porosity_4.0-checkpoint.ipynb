### 交叉验证模型评估（常规划分方式，10折）

```python
shuffle = ShuffleSplit(n_splits=10, test_size=9, random_state=42)
index = np.arange(46)
for train_index, test_index in shuffle.split(index):
    print(train_index)
    print(test_index)
```

```python
sns.set_style('darkgrid')
x2 = [i*0.02 for i in range(-1, 8)]
x2
```

```python
def cv_loop2(reg):
    mses = []
    maes = []
    fig = plt.figure(figsize=(10, 8))
    i = 1
    for train_index, test_index in shuffle.split(index):
        # 取出训练集和测试集数据
        x_train = new_df[['power', 'speed']].iloc[train_index]
        y_train = new_df['porosity'].iloc[train_index]
        x_test = new_df[['power', 'speed']].iloc[test_index]
        y_test = new_df['porosity'].iloc[test_index]
        # 模型的预测和训练
        reg.fit(x_train, y_train)
        print(reg.log_marginal_likelihood())
        y_predict = reg.predict(x_test)
        y_test_inverse = scaler_Y.inverse_transform(y_test)
        y_predict_inverse = scaler_Y.inverse_transform(y_predict)
        plt.scatter(y_test_inverse, y_predict_inverse)
        mse = mean_squared_error(y_test_inverse, y_predict_inverse)
        mae = mean_absolute_error(y_test_inverse, y_predict_inverse)
        mses.append(mse)
        maes.append(mae)
        i += 1
    plt.plot(x2, x2, 'r-')
    plt.xlim(-0.02, 0.12)
    plt.ylim(-0.02, 0.12)
    plt.xticks(x2)
    plt.yticks(x2)
    plt.xlabel('测试值', fontdict=font1)
    plt.ylabel('预测值', fontdict=font1)
    plt.title('孔隙率模型评估', fontdict=font1)
    plt.tick_params(labelsize=15)
    plt.show()
    return mses,maes
```

```python
kernel_1 = 1.0 * RBF(length_scale=10,\
                   length_scale_bounds=(1e-4, 1e4)) +\
WhiteKernel(noise_level=1e-3, noise_level_bounds=(1e-10, 1e+2))
reg_1 = GaussianProcessRegressor(kernel=kernel_1,\
                               n_restarts_optimizer=20,\
                               alpha=0, random_state=42)
```

```python
mses2, maes2 = cv_loop2(reg_1)
mses2
```

```python
np.mean(mses2), np.mean(maes2)
```

```python
ucb_res = pd.DataFrame(columns=['power', 'speed', 'porosity', 'ucb'])
temp = [temp1, temp2, temp3, temp4]
ucb = ['ucb-0.01', 'ucb-0.1', 'ucb-1', 'ucb-10']
k = 2
for t in temp:
    for u in ucb:
        sort_data = t.sort_values(by=u, ascending=True)
        filter_data = sort_data[['power', 'speed', 'porosity',\
                                 u]].iloc[0:k]
        index = [u, u]
        filter_data.index = index
        filter_data.columns = ucb_res.columns
        ucb_res = pd.concat([ucb_res, filter_data])
```
